```python
import tensorflow.keras as keras
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
#Change room ID to a numeric value
def room_to_numeric(cabin):

    return ord(cabin[0])
    
```

```python
all_df = pd.read_csv("train.csv")
all_df['Cabin'] = all_df['Cabin'].fillna('B')


all_df["Namevar"] = 0
all_df["moreroom"] = 0
all_df["name_length"] = 0
all_df["new_ticket"] =0
all_df["S"] = 0
all_df["C"] = 0
all_df["Q"] = 0
for j in range (0,len(all_df)):

    all_df.iloc[j,14]= len(all_df.iloc[j,3] )
    if("Mrs" in all_df.iloc[j,3]):
        all_df.iloc[j,12] = 0
    elif("Miss" in all_df.iloc[j,3]):
        all_df.iloc[j,12] = 1
    elif("Mr" in all_df.iloc[j,3]):
        all_df.iloc[j,12] = 2
    else:
        all_df.iloc[j,12] = 3

    if(' ' in all_df.iloc[j,10] ):
        all_df.iloc[j,13] = 1
    else:
        all_df.iloc[j,13] = 0

    all_df.iloc[j,10] = room_to_numeric(all_df.iloc[j,10] )
    if(all_df.iloc[j,4] == "female"):
        all_df.iloc[j,4] = 1
    else:
        all_df.iloc[j,4] = 0


    if(all_df.iloc[j,11] == 'S'):
        all_df.iloc[j,16] = 1
    elif(all_df.iloc[j,11] == 'C' ):
        all_df.iloc[j,17] = 1.5
    elif(all_df.iloc[j,11] == 'Q' ):
        all_df.iloc[j,18] = 1
    if('.' in all_df.iloc[j, 8] ):
        all_df.iloc[j,15] = 1
    else:
        all_df.iloc[j,15] = 0

```

```python
all_df[all_df["Embarked"].isna()]
```

```python



name_var_0 = all_df[all_df["Namevar"] == 0]["Age"].mean()
name_var_1 = all_df[all_df["Namevar"] == 1]["Age"].mean()
name_var_2 = all_df[all_df["Namevar"] == 2]["Age"].mean()
name_var_3 = all_df[all_df["Namevar"] == 3]["Age"].mean()


print(name_var_0, name_var_1, name_var_2)
all_df['Age'] = all_df['Age'].fillna(-99)
Na_age_L = all_df[all_df["Age"] == -99]

for i in range(0, len(Na_age_L)):


    if(Na_age_L.iloc[i,12] == 0):
        all_df.iloc[Na_age_L.iloc[i,0]-1 , 5] = name_var_0

    elif(Na_age_L.iloc[i,12] == 1):
        all_df.iloc[Na_age_L.iloc[i,0]-1 , 5] = name_var_1


    elif(Na_age_L.iloc[i,12] == 2):
        all_df.iloc[Na_age_L.iloc[i,0] -1, 5] = name_var_2

    else:
        all_df.iloc[Na_age_L.iloc[i,0]-1 , 5] = name_var_3

all_df = all_df.dropna()
```

```python
x_train = all_df.iloc[150:750,[2,4,7,9,12,14,15,16,17,18]]
x_test = all_df.iloc[0:150, [2,4,7,9,12,14,15,16,17,18]]

y_train = all_df.iloc[150:750,[1]]
y_test = all_df.iloc[0:150,[1]]
```

```python
x_train = x_train.to_numpy()
y_train = y_train.to_numpy()
x_train = np.asarray(x_train).astype(np.float32)
y_train = np.asanyarray(y_train).astype(np.float32)

x_test = x_test.to_numpy()
y_test = y_test.to_numpy()
x_test = np.asarray(x_test).astype(np.float32)
y_test = np.asarray(y_test).astype(np.float32)


x_train = tf.keras.utils.normalize(x_train, axis=1)
x_test = tf.keras.utils.normalize(x_test, axis=1)
```

```python
model = tf.keras.models.Sequential()
```

```python



model.add(tf.keras.layers.Dense(8, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(16, activation=tf.nn.relu))

model.add(tf.keras.layers.Dense(4, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(1, activation=tf.nn.sigmoid))
```

```python
model.compile(optimizer='ADAM',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python

es_callback = keras.callbacks.EarlyStopping(monitor='loss', patience=20)


model.fit(x_train, y_train, epochs=10000, verbose= 1, callbacks= [es_callback])
```

```python
val_loss, val_acc = model.evaluate(x_test, y_test)
```
